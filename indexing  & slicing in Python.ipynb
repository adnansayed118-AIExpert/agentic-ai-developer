# String Indexing and  Slicing
Indexing:- it  will return the position of value 
Slicing :- left index slicing will starts right from the start position whereas right index slicing starts from last position with n-1 formula

```python
s1='Welcometeam'
s1
```

```python
len(s1)
```

```python
type(s1)
```

```python
s1[5]
```

```python
s1[-4]
```

```python
s1[1:-4]
```

```python
s1[:] # it prints all string
```

```python
s1[7:10]
```

```python
s1[7:11]
```

```python
s1[6:8]
```

```python
s1[:-6]
```

```python
s1[1:10:2]
```

```python
s1[2:10]  # print from 2nd index to n-1 index that is 9 index 
```

```python
s1[4:] #print from 4th index till last index
```

```python
s1[0:]  #print from zero index till last
```

```python
s1[12:] 
```

```python
s1[:13]  #print all index starting from 0 to last if we slicing with left index with : and right index with  value
```

```python
s1[13] # print index out of range error as we call directly 13th index that is not available
```

```python
 s1[:16]   # print from zero to last avaiable index but wont give result for 16 
```

```python
s1[1:8:2] #forward slicing  #print from 1st index to n-1 that 8-1=7 index  and print step count 2 value only 
```

```python
s1[::3]  #count starts from zero to end index but print only step count 3 result from left indexing 
```

```python
s1[::2]             #count starts from zero to end index but print only step count 2 result from left indexing 
```

```python
s1[-10:-1:2] #backward slicing  #count starts from -10 to zero idex but print only step count 2 rsult from left indexing 
```

```python
s1[-9:-1:3]
```

```python
s1[::-1]  #backward step slicing print index from last index to zero index
```

```python
s1[::-3] 
```
