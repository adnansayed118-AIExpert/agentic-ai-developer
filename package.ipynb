# input() function

input() is a built-in Python function used to take user input from the keyboard.

It always returns the input as a string (str).

note:- input() by default understand string
if user wanted to convert string to int () use typecasting
if user wanted to slicing use string slicing
if user to wanted to perform mathematical operation in input() fucntion the use "eval()" function

syntax:-

input("Message to display")



```python
# it will not add x+y values instead it will merge x and y values as input () will only read and return string values 

x = input()
y = input()
z = x + y
print (z)

```

```python
# in order to get addition values use int()
x = int(input())
y = int(input())
z = x + y
print (z)
```
