# print with separator

here  we use only print statement and under one print () , will pass multipple values and can separate these multi values with any thing

```python
print ('hi','my name is','adnan','living in','sharjah UAE', sep= '-->')
```

```python
print ('hi','my name is','adnan','living in','sharjah UAE', sep= '@')
```

```python
print ('hi','my name is','adnan','living in','sharjah UAE', sep= '$')
```

```python
print(1,2, end = ' ')
print (3, '.',  sep='')
```

```python
print(1,2, end = ' ')
print (3, 'hree',  sep='')
```
