# Tuple

#  A tuple is an ordered, immutable collection that allows duplicates.

🔹 Features:

Ordered ✅

Immutable ❌ (cannot change)

Allows duplicates ✅

Indexed ✅

🔹 Example:

my_tuple = (10, 20, 30, 10)



👉 Used when data should not change (e.g., coordinates, fixed values).

```python
tup1=()  # empty tuple
```

```python
tup1
```

```python
tup2=(10,30,60)  #tuple with int value
tup2
```

```python
tup3=(10.77,30.66,60.89) # tuple with float value
tup3
```

```python
tup4=('one','two','three')   #tuple of strings
tup4
```

```python
tup5=('Asif',25,(50,100),(150,90)) #tuple of nested tuple
tup5
```

# tuple slicing

```python
mytuple=('one','two','three','four','five','six','seven','eight')
mytuple
```

```python
mytuple[0:3]   # start with index 0 and end with index 3-1 (n-1)
```

```python
mytuple[2:5]    #list of all items from 2 to 5 with n-1 end index (5-1)
```

```python
mytuple[:3]   # return first three items
```

```python
mytuple[0:2] # return first two items
```

```python
mytuple[-3:]  #return last 3 items
```

```python
mytuple[-2:]  #return last 2 items
```

```python
mytuple[-1:]  #return last  item of the tuple
```

```python
del mytuple[0]   # tuples are immutable which means we cannot delete the tuple items
```

```python
mytuple[0]=1       # tuples are immutable which means we cannot change the tuple items
```

```python
del mytuple   # deleting entire tuple object is possible 
```

```python
mytuple
```

# Loop through tuple

```python
mytuple=('one','two','three','four','five','six','seven','eight')
mytuple
```

```python
for i in mytuple:
    print(i)
```

```python
for i in enumerate(mytuple):
    print(i)
```

# tuple member ship

```python
'one' in mytuple  # check if 'one' exists in the tuple
```

```python
'ten' in mytuple  # check if 'ten' exists in the tuple
```

```python
if 'three' in mytuple:              # check if 'three' exists in the tuple
    print("three is available") 
else:
    print ("three is not available")
```

```python
if 'ten' in mytuple:      # check if 'ten' exists in the tuple
    print("ten is available")
else:
    print ("ten is not available")
```
