# 5. left shift <<

in left shift << operator we gain the bit

```python
12 << 1  #bin(12) ob 1100 << 1 == 11000
```

```python
12 << 2 #bin(12) ob 1100 << 2 == 110000
```

```python
12 << 3 #bin(12) ob 1100 << 3 == 1100000
```

# 6. right shift >> 

In right shift operator we loose the bit

```python
12>>2 # bin of 12  is 1100 , so if applying right shift >> then 2 bits will be loose and the result will be 11
```

```python
12>>3 # bin of 12  is 1100 , so if applying right shift >> then 3 bits will be loose and the result will be 1
```

```python
bin (15)
```

```python
15 >> 2 # bin of 15  is 1111 , so if applying right shift >> then 2 bits will be loose and the result will be 11
```
